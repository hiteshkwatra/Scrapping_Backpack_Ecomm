## Scrapping  the backpack data with discount if available for FlipKart Ecommerce Website

```python
import requests as req
from bs4 import BeautifulSoup as bfs
```

```python
bags=[]
discount=[]
listOfPrices=[]

for i in range(1,9):
    
    url="https://www.flipkart.com/search?as=off&as-show=on&otracker=start&page=" + str(i) + "&q=bags&viewType=grid"
    r= req.get(url)
    soup= bfs(r.content)
    print(i)
    bagsdata=soup.find_all("div",{"class":"_2SxMvQ"})
    
    for item in bagsdata:
        for i in range(len(item.contents)):
            for j in range(len(item.contents[i].find_all("a", {"class":"_2cLu-l"}))):
                           
                bags.append(item.contents[i].find_all("a", {"class":"_2cLu-l"})[j].text)
                
                listOfPrices.append(item.contents[i].find_all("div",{"class":"_1vC4OE"})[j].text)
                
                
                try:
                    discount.append(item.contents[i].find_all("div",{"class":"VGWI6T"})[j].text)
                except:
                    discount.append("no discount")
                
    
   
        
       
        
        

        
print(len(bags),len(listOfPrices),len(discount))
```

```python
for i in range(320):
    print(i,bags[i],listOfPrices[i],discount[i])
```

## Putting the data into firebase database and retrieving it back

```python
from firebase import firebase
firebase = firebase.FirebaseApplication('https://first-firebase-project-6e849.firebaseio.com/', None)
```

```python
for i in range(320):
    productName=bags[i]
    result=firebase.put_async('/bags',productName,{"index": i,"Price":listOfPrices[i],"Discount_Available":discount[i]})
    print(result)
```
